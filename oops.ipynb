```python
## 3. Explain the diffrence between instance methods and class methods. Provide an example of each
Instance Methods:
1.Bound to an instance of the class and can access/modify instance attributes.
2.First parameter is self, which refers to the instance.
3.No special decorator is required.
##Example-
class Dog:
    def __init__(self, name):
        self.name = name

    def show_name(self):  # Instance method
        return f"My dog's name is {self.name}."

my_dog = Dog("Buddy")
print(my_dog.show_name())  # Output: My dog's name is Buddy.

Class Methods:
1.Bound to the class itself, and can access/modify class-level attributes (shared across instances).
2.First parameter is cls, which refers to the class.
3.Uses the @classmethod decorator.
##Example-
class Dog:
    species = "Canine"  # Class attribute

    @classmethod
    def show_species(cls):  # Class method
        return f"All dogs are {cls.species}."

print(Dog.show_species())  # Output: All dogs are Canine.

```

```python
## 4. How does python implement method overloading. Give an example

Python does not support method overloading natively (multiple methods with the same name but different parameters).
Instead, it uses techniques like default arguments or variable-length arguments (*args) to simulate overloading.

##Example-
class Calculator:
    def add(self, a, b, c=0):  # Default argument c=0
        return a + b + c

calc = Calculator()
print(calc.add(5, 10))      # Output: 15 (adds 2 numbers)
print(calc.add(1, 2, 3))    # Output: 6 (adds 3 numbers)
```

```python
## 5. What are the three types of access modifiers in python.How are they denoted

In Python, there are three types of access modifiers that control the visibility of class members (attributes and methods).
 While Python doesn’t have strict access modifiers like some other languages (e.g., private, protected, public),
 it follows naming conventions to simulate them.

 ##1. Public-
1. Members that are accessible from anywhere (inside and outside the class).
2.as_integer_ratioDenoted by no underscores before the name.
##Example-
class Car:
    def __init__(self, make, model):
        self.make = make  # Public attribute
        self.model = model  # Public attribute

car = Car("Toyota", "Corolla")
print(car.make)  # Accessible

## 2. Protected-
1. Members that should be accessible within the class and its subclasses, but not intended for external use.
2. Denoted by a single underscore (_) before the name.
## Example-
class Car:
    def __init__(self, make, model):
        self._make = make  # Protected attribute
        self._model = model  # Protected attribute

car = Car("Toyota", "Corolla")
print(car._make)  # Accessible, but conventionally "protected"

## 3. Private-
1.Members that are intended to be accessible only within the class.
2.Denoted by a double underscore (__) before the name.
3.Python performs name-mangling to make the attribute inaccessible from outside the class,
but it can still be accessed using special syntax.

##Example-
class Car:
    def __init__(self, make, model):
        self.__make = make  # Private attribute

    def get_make(self):  # Public method to access private attribute
        return self.__make

car = Car("Toyota", "Corolla")
print(car.get_make())  # Accessed via public method
# print(car.__make)    # Raises an AttributeError



```

```python
## 6. Describe 5 types of inheritance in Python. Provide a simple example of multiple inheritance.
1.Single Inheritance-
In single inheritance, a child class inherits from a single parent class. This is the simplest form of inheritance.

2.Multiple Inheritance-
In multiple inheritance, a child class inherits from more than one parent class.
This means that the child class can access attributes and methods of all the parent classes.

3.Multilevel Inheritance-
In multilevel inheritance, a class inherits from a parent class, which in turn inherits from another parent class.
This forms a chain of inheritance.

4.Hierarchical Inheritance-
In hierarchical inheritance, multiple child classes inherit from the same parent class. This allows different subclasses to share common attributes and methods from a single parent.

5. Hybrid Inheritance-
Hybrid inheritance is a combination of two or more types of inheritance. It may involve multiple and hierarchical inheritance structures,
allowing more complex relationships between classes.

##Example of Multiple Inheritance-
# Parent class 1
class Parent1:
    def greet(self):
        print("Hello from Parent1")

# Parent class 2
class Parent2:
    def greet2(self):
        print("Hello from Parent2")

# Child class inheriting from both Parent1 and Parent2
class Child(Parent1, Parent2):
    pass

# Creating an instance of the Child class
child = Child()

# Calling methods from both parent classes
child.greet()   # Output: Hello from Parent1
child.greet2()  # Output: Hello from Parent2




```

```python
## 7. What is the method resolution order in Python? How can you retrive it programatically?
The Method Resolution Order (MRO) in Python is the order in which Python looks for a method or attribute in a hierarchy of classes during inheritance. This is especially important in cases of multiple or hybrid inheritance, where a class inherits from more than one parent.

Key Aspects of MRO:
1.Order of Search: When a method is called on an object, Python looks for the method in the class of that object.
If it's not found, it checks the parent classes in a specific order, determined by the MRO.

2.Depth-First Search: Python's MRO follows a left-to-right depth-first search in the class hierarchy.

3.Linearization Using C3 Algorithm: Python uses the C3 linearization algorithm to determine the MRO.
This ensures a consistent, logical order of method resolution when there is multiple inheritance.

4.super(): The super() function uses the MRO to determine which method to call next, based on the inheritance chain.

##We can retrive it programatically as follows:
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass

# Using __mro__ attribute
print(D.__mro__)
# Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)

# Using mro() method
print(D.mro())
# Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```

```python
## 8. Create an abrstract base class 'Shape' with an abstract method 'area()'.Then create two subclasses 'Circle' and 'Rectangle' that implement the 'area' method.
from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Subclass for Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of Circle: {circle.area()}")  # Output: Area of Circle: 78.53981633974483
print(f"Area of Rectangle: {rectangle.area()}")  # Output: Area of Rectangle: 24

```

```python
## 9. Demostrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.
from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Subclass for Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

# Subclass for Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Function that demonstrates polymorphism
def print_area(shape):
    print(f"The area of the {shape.__class__.__name__} is: {shape.area()}")

# Create instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Use the same function to calculate and print the area of both shapes
print_area(circle)       # Output: The area of the Circle is: 78.53981633974483
print_area(rectangle)    # Output: The area of the Rectangle is: 24

Polymorphism allows functions to operate on different objects in a way that is agnostic to their specific types,
as long as these objects implement the same method or interface.
 In this case, since both Circle and Rectangle are subclasses of Shape and implement the area() method,
 we can create a function that works with any Shape object to calculate and print the area.

 Polymorphic Function (print_area):
This function takes a shape object as an argument and calls the area() method on it.
The function doesn’t need to know whether the shape is a Circle, Rectangle, or any other subclass of Shape,
as long as the shape object implements the area() method.

How Polymorphism Works:
The print_area() function works polymorphically. It can take either a Circle object or a Rectangle object (or any other subclass of Shape that implements area()).
The __class__.__name__ is used to print the name of the specific class (e.g., Circle or Rectangle).

Function Usage:
We create instances of both Circle and Rectangle, and then call print_area() with each.
Python automatically knows which area() method to call based on the object's actual class.
```

```python
## 10. Implement encapsulation in a 'BankAccount' class with private attributes for 'balance' and 'account_number'. Include methods for deposit, withdrawal and balance inquiry.
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute for account number
        self.__balance = initial_balance        # Private attribute for balance

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew: ${amount}")
            else:
                print("Insufficient balance.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to check the balance
    def get_balance(self):
        return self.__balance

    # Method to get the account number (if needed)
    def get_account_number(self):
        return self.__account_number


# Example usage
account = BankAccount("1234567890", 500)  # Create an account with $500 initial balance

# Deposit money
account.deposit(200)      # Output: Deposited: $200

# Withdraw money
account.withdraw(100)     # Output: Withdrew: $100
account.withdraw(700)     # Output: Insufficient balance.

# Check balance
print(f"Balance: ${account.get_balance()}")  # Output: Balance: $600

# Access account number (if needed)
print(f"Account Number: {account.get_account_number()}")  # Output: Account Number: 1234567890

```

```python
## 11. Write a class that overrides the '__str__' and '__add__' magic methods. What will these methods allow you to do?
class MyNumber:
    def __init__(self, value):
        self.value = value

    # Override __str__ to provide a readable string representation
    def __str__(self):
        return f"MyNumber: {self.value}"

    # Override __add__ to define how + operator works between two MyNumber objects
    def __add__(self, other):
        if isinstance(other, MyNumber):
            return MyNumber(self.value + other.value)
        return NotImplemented

# Example usage
num1 = MyNumber(10)
num2 = MyNumber(20)

# Demonstrate __str__ method
print(num1)  # Output: MyNumber: 10
print(num2)  # Output: MyNumber: 20

# Demonstrate __add__ method
result = num1 + num2
print(result)  # Output: MyNumber: 30

These methods allow you to do:

__str__:
This method is called when you use str() or print() on an object of the class. It allows you to define how the object should be represented as a string.
In this example, print(num1) will output MyNumber: 10 instead of the default object representation (<__main__.MyNumber object at ...>), which is usually not very readable.

__add__:
This method allows you to define how the + operator works for objects of the class.
In the example, num1 + num2 adds the value of both MyNumber objects and returns a new MyNumber object with the sum of the values (30 in this case).
Without overriding __add__, Python would not know how to add two MyNumber objects and would raise an error.
```

```python
## 12. Create a decorator that measures and prints the execution time of function
import time

# Decorator to measure and print the execution time of a function
def execution_time_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record the start time
        result = func(*args, **kwargs)  # Call the actual function
        end_time = time.time()  # Record the end time
        execution_time = end_time - start_time  # Calculate execution time
        print(f"Execution time of {func.__name__}: {execution_time:.4f} seconds")
        return result
    return wrapper

# Example usage of the decorator
@execution_time_decorator
def example_function():
    time.sleep(2)  # Simulate a function that takes some time to execute
    print("Function execution complete")

# Call the decorated function
example_function()

```

```python
## 13.. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?
The Diamond Problem is an issue that arises in multiple inheritance when a class inherits from two or more classes that have a common ancestor, creating an ambiguity in the inheritance chain.
 The problem is named because the inheritance diagram often forms a diamond shape

class A:
    def greet(self):
        print("Hello from A")

class B(A):
    def greet(self):
        print("Hello from B")

class C(A):
    def greet(self):
        print("Hello from C")

class D(B, C):
    pass

d = D()
d.greet()

How Python Resolves the Diamond Problem:
Python uses a method resolution strategy called C3 Linearization (also known as the C3 superclass linearization algorithm) to resolve this issue.
 This ensures that the method resolution order (MRO) is consistent, predictable, and avoids duplication of methods when traversing the inheritance chain.

In Python, the MRO (Method Resolution Order) defines the order in which methods are looked up in a class hierarchy.
 This order is calculated when the class is defined and can be viewed using the __mro__ attribute or the mro() method.

```

```python
## 14.  Write a class method that keeps track of the number of instances created from a class.
class InstanceCounter:
    # Class variable to keep track of the number of instances
    instance_count = 0

    def __init__(self):
        # Increment the class variable each time a new instance is created
        InstanceCounter.instance_count += 1

    # Class method to return the current instance count
    @classmethod
    def get_instance_count(cls):
        return cls.instance_count


# Example usage
a = InstanceCounter()
b = InstanceCounter()
c = InstanceCounter()

# Calling the class method to get the instance count
print(f"Number of instances created: {InstanceCounter.get_instance_count()}")  # Output: 3

```

```python
## 15. Implement a static method in a class that checks if a given year is a leap year.
class DateUtils:

    @staticmethod
    def is_leap_year(year):
        # A leap year is divisible by 4, but not divisible by 100, unless it's divisible by 400
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        return False

# Example usage:
print(DateUtils.is_leap_year(2020))  # Output: True
print(DateUtils.is_leap_year(2021))  # Output: False
print(DateUtils.is_leap_year(2000))  # Output: True
print(DateUtils.is_leap_year(1900))  # Output: False
```
